```python
import os

yes_images = []
no_images = []
image_extensions = ".jpg"

nins_dir = os.path.join(path, "NINS_Dataset", "NINS_Dataset")
for subfolder in os.listdir(nins_dir):
    subfolder_path = os.path.join(nins_dir, subfolder)
    if not os.path.isdir(subfolder_path):
        continue

    image_files = []
    for root, dirs, files in os.walk(subfolder_path):
        for file in files:
            if file.lower().endswith(image_extensions):
                image_files.append(os.path.join(root, file))

    if subfolder.lower() == "normal":
        no_images.extend(image_files)
    else:
        yes_images.extend(image_files)

print(f"{len(yes_images)} images for YES")
print(f"{len(no_images)} images for NO")
```

```python
import matplotlib.pyplot as plt
import cv2
import random

def show_samples(image_list, label, num=5):
    plt.figure(figsize=(15, 5))
    for i, img_path in enumerate(random.sample(image_list, num)):
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.subplot(1, num, i+1)
        plt.imshow(img)
        plt.title(label)
        plt.axis('off')
    plt.show()

print("YES class samples:")
show_samples(yes_images, "Tumor")

print("NO class samples:")
show_samples(no_images, "Normal")
```

```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.losses import BinaryCrossentropy
import kagglehub

path = kagglehub.dataset_download("sudipde25/mri-dataset-for-detection-and-analysis")

yes_images = []
no_images = []

nins_dir = os.path.join(path, "NINS_Dataset", "NINS_Dataset")

tumor_folders = [
    'Brain tumor - Recurrenceremnant of previous lesion',
    'Brain tumor operated with ventricular hemorrhage',
    'Brain Tumor',
    'Brain Tumor (Hemangioblastoma  Pleomorphic xanthroastrocytoma  metastasis)',
    'Brain tumor (Dermoid cyst craniopharyngioma)',
    'small meningioma',
    'pituitary tumor',
    'Brain Tumor (Ependymoma)',
    'meningioma',
    'Brain tumor (Astrocytoma Ganglioglioma)',
    'Glioma',
    'Left Retro-orbital Haemangioma'
]
normal_folder = 'Normal'


for folder in tumor_folders:
    folder_path = os.path.join(nins_dir, folder)
    if os.path.isdir(folder_path):
        for fname in os.listdir(folder_path):
            if fname.lower().endswith(".jpg"):
                yes_images.append(os.path.join(folder_path, fname))


normal_path = os.path.join(nins_dir, normal_folder)
for fname in os.listdir(normal_path):
    if fname.lower().endswith(".jpg"):
        no_images.append(os.path.join(normal_path, fname))


# combine yess and nos
data = []
labels = []

image_size = 150  # resize to 150x150

for img_path in yes_images:
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img, (image_size, image_size))
    img = img / 255.0
    data.append(img)
    labels.append(1) #yes=1

for img_path in no_images:
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img, (image_size, image_size))
    img = img / 255.0
    data.append(img)
    labels.append(0) #no=0


data = np.array(data).reshape(-1, 150, 150, 1)
labels = np.array(labels)

# shuffle
indices = np.arange(len(data))
np.random.shuffle(indices)

data = data[indices]
labels = labels[indices]

# split the data into 80:20
x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=10)


model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), padding='same', activation='relu', input_shape=(150, 150, 1)),  # 32x150x150
    tf.keras.layers.MaxPooling2D(pool_size=(2,2)),  # 32x75x75

    tf.keras.layers.Conv2D(64, (3,3), padding='same', activation='relu'),  # 64x75x75
    tf.keras.layers.MaxPooling2D(pool_size=(2,2)),  # 64x37x37

    tf.keras.layers.Conv2D(128, (3,3), padding='same', activation='relu'),  # 128x37x37
    tf.keras.layers.MaxPooling2D(pool_size=(2,2)),  # 128x18x18

    tf.keras.layers.Conv2D(128, (3,3), padding='same', activation='relu'),  # 128x18x18
    tf.keras.layers.MaxPooling2D(pool_size=(2,2)),  # 128x9x9

    tf.keras.layers.Flatten(),

    # one hidden layers consisting of 128 neurons
    # then the output layer

    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(1, activation='sigmoid')  # sigmoid output
])

model.compile(optimizer='adam',
              loss=BinaryCrossentropy(),
              metrics=['accuracy'])

epochs = 15
history = model.fit(x_train, y_train, epochs=epochs, batch_size=25)

# plotting accuracy and loss curves

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label="loss")
plt.xlabel("epochs")
plt.ylabel("loss")
plt.title("training loss")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label="accuracy")
plt.xlabel("epochs")
plt.ylabel("accuracy (%)")
plt.title("training accuracy")
plt.legend()

plt.show()

# final evaluation
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f"test Accuracy: {test_accuracy * 100:.2f}%")
model.summary()
```
